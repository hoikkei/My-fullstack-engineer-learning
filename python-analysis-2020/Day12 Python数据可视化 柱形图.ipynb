绘制柱形图主要使用matplotlib的bar()函数，bar()的构造函数：bar(x,height,width,*,align='center',**kwargs)

x:包含所有柱子的下标的列表

height:包含所有柱子的 高度值的列表

width: 每个柱子的高度。可以指定一个固定值，那么所有的柱子都是一样的宽。 或者设置一个列表，这样可以分别对每个柱子设定不同的宽度。

align: 柱子对齐方式，有两个可选值：center和edge.center表示每根柱子是根据下标来对齐，edge则表示每根柱子全部以下标为起点，然后显示到下标的右边。
如果不指定该参数，默认值是center。

color: 每根柱子呈现的颜色。同样可指定一个颜色值，让所有柱子呈现同样颜色；或者指定带有不同颜色的列表，让不同柱子显示不同的颜色。

edgecolor:每根柱子边框的颜色。同样可指定一个颜色值，让所有柱子边框呈现同样颜色；或者指定带有不同颜色的列表，让不同柱子的边框显示不同颜色。



## （1） 绘制柱形图

```python
import numpy as np
import matplotlib.pyplot as plt
data = [23,58,74,67,87]
plt.bar([1,2,3,4,5],data)
```

(2) 设置颜色

color 参数可以设置颜色，alpha参数可以设置透明度


```python
plt.bar([1,2,3,4,5],data,color='green',alpha=0.7)
```

```python
plt.bar([1,2,3,4,5],data,color='rgby',alpha=0.7)
```

(3) 添加坐标轴


```python
data = [23,85,72,43,53]
labels=['a','b','c','d','e']
plt.xticks(range(len(data)),labels)
plt.xlabel('x')
plt.ylabel('y')
plt.bar(range(len(data)),data)
```

(4) 设置刻度和刻度标签



```python
data = [23,85,72,43,53]
labels=['a','b','c','d','e']
plt.xticks(range(len(data)),labels)
plt.bar(range(len(data)),data)
```

（5） 网格设置

```python
plt.bar([1,2,3,4,5],data,color='green',alpha=0.7)
plt.grid(color="#95a5a6",linestyle='--',linewidth=1,axis='y',alpha=0.5)
```

(6) 水平柱状图

```python
data1 =[23,85,72,43,53]
plt.barh(range(len(data1)),data1)
```

(7) 并列柱状图

```python
data1=[23,85,72,43,53]
data2=[42,35,21,16,13]
plt.bar(range(len(data1)),data1,alpha=0.7)
plt.bar(range(len(data2)),data2,bottom=data1,alpha=0.7)
```

（8）堆积柱状图

```python
data1=[23,85,72,43,53]
data2=[42,35,21,16,13]
width=0.3
plt.bar(np.arange(len(data1)),data1,width=width,label='blue')
plt.bar(np.arange(len(data2))+width,data2,width=width,label='yellow')
plt.legend()
```
