1. 数据的提取

.head()函数：默认读取数据的前5行

首先，导入科学技术库Numpy和数据分析库Pandas,读取一数据集"商场客户信息" ,用"data" 来存储，并利用.head() 函数查看数据的前五行，保证读取无误



```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("商场客户信息.csv")
data.head()
```

.tail()函数 ：读取数据的后5行

```python
data.tail()
```

iloc 函数
用于提取指定行，比如 ：提取数据中的第1行、第4行、第5行、第6行

是一种能够按照位置对数据进行提取的一种函数，它的索引不再是一些标签，而是一些从0开始的索引值

```python
data.iloc[[0,3,4,5]]
```

用列索引（变量名）提取指定名称

```python
data[['CustomerID','Gender','Age']].head()
```

```python
data.loc[[0,4,5,6],['CustomerID','Gender','Age']]
```

3、通过布尔索引实现查询

和在Numpy中实现布尔索引的用法相似，如果想要获取所有“性别”为“女”的数据信息

```python
data[data['Gender']=='Female']
```

在此基础上，如果想要获取50岁以上的女性信息

```python
data[(data['Gender']=='Female')&(data['Age']>50)]
```

```python
data[(data['Gender']=='Female')&(data['Age']>50)][['Annual Income (k$)','Spending Score (1-100)']]
```
