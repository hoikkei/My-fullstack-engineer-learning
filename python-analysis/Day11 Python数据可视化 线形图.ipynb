## 线形图

线形图是最基本的图表类型，常用于绘制连续的数据。通过绘制线形图，可以表现出数据的趋势。

绘制图形主要用到两个库，matplotlib.pyplot 和numpy


```python
# 导入numpy包并简写为np,导入matplotlib.pyplot 并简写为plt

import numpy as np
import matplotlib.pyplot as plt
```

#### (1) 绘制直线 y=3*x

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,50)   # 从（-1，1）均匀取50个点
y = 3 * x
plt.figure()
plt.plot(x,y)
plt.show()
```

#### 设置横纵轴坐标


```python
# 设置横纵坐标轴标签

x = np.linspace(-1,1,50)   # 从（-1，1）均匀取50个点
y = 3 * x
plt.figure()
plt.plot(x,y)

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 设置横纵坐标上下限
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,50)   # 从（-1，1）均匀取50个点
y = 3 * x
plt.figure()
plt.plot(x,y)

plt.xlabel("X")
plt.ylabel("Y")

plt.xlim(-0.75,0.75)
plt.ylim(-2,2)
plt.show()
```

```python
# 设置横轴坐标精准刻度
x = np.linspace(-1,1,50)   # 从（-1，1）均匀取50个点
y = 3 * x
plt.figure()
plt.plot(x,y)

plt.xticks([-1,-0.5,0,0.5,1])
plt.yticks([-2,-1,0,1,2])
plt.show([-2,-1,0,1,2])
```

(9) 注释特殊点

scatter() han函数必须含两个参数x和y.它们的数据类型为列表。x代表要标注点的横轴位置，y代表要标注点的横轴位置。x和y列表中下标相同的数据是对应的。
例如x为[3,4] ,y为[6,8].因此，x要和y长度要一样。

annotate()函数同样也必须有两个参数，一个是标注内容，另一个是xy。标注内容是一个字符串。xy表示要在哪一个位置（点）显示标注内容

点被标注出来后，如果需要添加注释，使用text()函数。text(x,y,s)作用是在点（x,y)上添加文本s。fontdict是对注释的字体的颜色进行设置。



```python
# 将重要的点标出并做标注

plt.plot(x,y)
plt.scatter([0],[0],s=10,color='g')
plt.annotate('(0,0)',xy=(0,0),fontsize=20)
plt.text(0,1,"this point very import",fontdict = {"size":12,"color":"green"})
```
