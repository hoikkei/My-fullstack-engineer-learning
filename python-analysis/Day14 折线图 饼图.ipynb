```python
import pandas as pd
unrate=pd.read_excel("销售量.xlsx")
unrate['date']=pd.to_datetime(unrate['date'])
print(unrate.head(10))
```

```python
import matplotlib.pyplot as plt
first_twelve=unrate[0:12]
second_twelve=unrate[0:12]
#指定画布大小
fig=plt.figure(figsize=(6,3))
#绘制两条折线
plt.plot(first_twelve['date'],first_twelve['value1'],color='blue')
plt.plot(second_twelve['date'],second_twelve['value2'],color='red')
plt.xticks(rotation=45)  #旋转x轴刻度
plt.xlabel("June")
plt.ylabel("sales")
legend_text = ['value1','value2']
plt.legend(legend_text,loc='upper left')
plt.title("销售量")
plt.show()

```

plt.pie(x,explode=None,labels=None,colors=None,autopct=None,pctdistance=0.6,shadow=False,labeldistance=1.1,startangle=None,radius=None,counterclock=True,wedgeprops=None,textprops=None,center=(0,0),frame=False,rotatelabels=False,hold=None,data=None)

x：表示用于绘制饼图的数据。
explode：用来指定每部分的偏移量或者说指定项距离饼图圆心为n个半径。
labels：指定每一个饼块的名称。
colors：表示饼图颜色
autopct：指定数值的显示方式。
autopct: 设置百分数比例显示格式，如%3.1f%%表示3位有效数字，保留小数点后一位，浮点类型，%%为了显示出百分号
pctdistance：默认值：0.6,设置百分数字标签离中心距离。
labeldistance：默认值：1.1,指定每一项的名称label和距离圆心的半径。
startangle：从x轴逆时针旋转饼图的开始角度。
radius：饼图的半径。
wedgeprops:设置饼形图内外边界的属性，如环的宽度，环边界颜色和宽度

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
#生成数据
labels = ["A","B","C","D","其他"]
x = [0.45,0.25,0.15,0.05,0.10]
explode = [0.1,0,0,0,0]  #指定每块饼干边缘偏离半径的百分比
autopct = '%3.1f%%'  #指定数值的显示方式
startangle = 180  #从x轴逆时针旋转饼图的开始角度
shadow = True  #在饼图下面画一个阴影
radius = 1     #饼图的半径
labeldistance=1.1 #每一项的名称label和距离圆心的半径
plt.pie(x,explode = explode,labels = labels,autopct = '%3.1f%%',
        startangle =180, shadow =True,radius=1,labeldistance=1,
        center=(0,0),colors = ['c','r','gray','g','y'])
plt.title("手机销量")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']

x1=[20,30,15,20,15]
x2=[30,25,15,20,10]

# 设置饼状图各个区块的颜色
color=['aqua','linen','lightcoral','olive','gold']
plt.pie(x1,autopct="%3.1f%%",radius=1,pctdistance=0.85,
        colors=color,wedgeprops=dict(linewidth=2,width=0.3,edgecolor='w'))
plt.pie(x1,autopct="%3.1f%%",radius=0.7,pctdistance=0.7,
        colors=color,wedgeprops=dict(linewidth=2,width=0.4,edgecolor='w'))

#图例
legend_text = ['萝莉',"屌丝","御姐","绿茶","帅哥"]
plt.legend(legend_text,title='画像',loc='center right')
plt.axis("equal")  #设置坐标轴比例以显示为圆形
plt.title("性格画像比例图")
plt.show()
```
