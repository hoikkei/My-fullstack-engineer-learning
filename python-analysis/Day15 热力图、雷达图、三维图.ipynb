热力图是将某事物的响应程度反映在图表上，可以快速发现需重点关注的区域，使用的plt库中的imshow方法，较多用于分析网站访客热衷的页面区域和访客所在的地理区域的图示。由此我们可将热力图分为线上&线下，线上的热力图其实根据用户点击次数来绘制，其目的就是给网站运营&移动运营产品经理等提供对页面按钮&功能参考，从而得知用户兴趣倾向，精简按钮等，比较广泛应用在在网站、App的用户分析中。

```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
cm=np.array([[1,0.082,0.031,-0.0086],[0.082,1,-0.063,0.062],[0.031,-0.09,1,0.026],[-0.0086,0.062,0.026,1]])
#设置配色方案
cmap=plt.cm.Spectral
plt.imshow(cm,cmap=cmap)
#显示右边的颜色条
plt.colorbar()
#防止中文乱码
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']= False
#设置x轴和y轴的刻度标签
classes=['负债率','信贷数量','年龄','家属数量']
tick_marks=np.arange(len(classes))
plt.xticks(tick_marks,classes)
#将数值显示在指定位置
plt.yticks(tick_marks,classes)
for i,j in itertools.product(range (cm.shape [0]),range(cm.shape[1])):
    plt.text(j,i,cm[i,j],horizontalalignment='center')
#设置网格线
plt.grid(False)
plt.title('热力图')
```

```python
import numpy as np

import pandas as pd 

import matplotlib.pyplot as plt

plt.subplot(111,polar=True)#建立一个坐标系,其中polar=True表示建立一个极坐标系

dataLenth=5#表示把整个圆分成5份

#np.linspace表示指定的间隔内返回均匀间隔的数字

angles=np.linspace(0,2*np.pi,dataLenth,endpoint=False)

labels=['沟通能力','业务理解能力','逻辑思维能力','快速学习能力','工具使用能力']

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

data=[2,3.5,4,4.5,5]

data=np.concatenate((data,[data[0]]))#闭合

angles=np.concatenate((angles,[angles[0]]))#闭合

#绘图

plt.polar(angles,data,color='b',marker='o')#绘图

plt.title('某数据分析师的综合评级')#设置标题名

plt.xticks(angles,labels)#设置x轴刻度
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()

plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

ax=Axes3D(fig)#画三维图

x=np.arange(-4,4,0.5)#给定x值

y=np.arange(-4,4,0.5)#给定y值

x,y=np.meshgrid(x,y)

z=np.sqrt(x**2+y**2)#计算z值

ax.plot_surface(x,y,z,color='r')#绘制三维曲面图
```
